```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#load titanic dataset
data = sns.load_dataset('titanic')
data.head()
```

```python
# check the number of missing values in each column
data.isnull().sum().sort_values(ascending=False)
```

```python
#drop deck column
data.drop('deck', axis=1, inplace=True)
data.head()
```

```python
#input missing values with meand
data['age'] = data['age'].fillna(data['age'].mean())
#check the number of missing values in each column
data.isnull().sum().sort_values(ascending=False)
```

```python
#input missing values with meand
data['age'] = data['age'].fillna(data['age'].median())
#check the number of missing values in each column
data.isnull().sum().sort_values(ascending=False)
```

```python
# input missing values with mode
data['embark_town']=data['embark_town'].fillna(data['embark_town'].mode())
#check the number of missing values in each column
data.isnull().sum().sort_values(ascending=False)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=4)
data['age'] = imputer.fit_transform(data[['age']])
data.isnull().sum().sort_values(ascending=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
data = sns.load_dataset('titanic')
data.head()
```

```python
#check the number of missing values in each column
data.isnull().sum().sort_values(ascending=False)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
#CREATE A LABELENCODER object using LabelEncoder() in for loop for categorical columns
# columns to encode
columns_to_encode = ['sex', 'embarked', 'who', 'deck', 'class', 'embark_town', 'alive']
#dictionary to store labelencoder for each column
label_encoder = {}
for col in columns_to_encode:
    #create a new labelencoder for the column
    le = LabelEncoder()
    # fit and transform the data
    data[col] = le.fit_transform(data[col])
    #store the encoder in the dictionary
    label_encoder[col] = le
    data.head()
```

```python
data.head()
```

```python
#input the missing values with iterative imputer
#call the iterative imputer class with  max iter=10
imputer = IterativeImputer(max_iter=5)
#input missing values using iterative imputer in a for loop for age, embark town, embarked columns
#columns to input
columns_to_impute = ['age', 'embark_town', 'embarked', 'deck']
#loop to input each column
for col in columns_to_impute:
    data[col] = imputer.fit_transform(data[[col]])
    #check the missing values
    data.isnull().sum().sort_values(ascending=False)
```

```python
data.head()
```

```python
#inverse transform
for col in columns_to_encode:
    le = LabelEncoder()
    data[col] = label_encoder[col].inverse_transform(data[col].astype(int))
    data.head()
```
